*Data Cleaning and Transformation*

**1. Importing necessary Python Libraries** 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandoc as pn
```

**2. Opening the csv file in Jupyter Notebook**

```python
df_sales_data = pd.read_csv("C:/Users/HP/Downloads/Python_Diwali_Sales_Analysis/Python_Diwali_Sales_Analysis/Diwali Sales Data.csv", encoding = "unicode_escape")
df_sales_data
```

```python
df_sales_data.shape
```

**3. Getting the overview of the dataframe**

```python
df_sales_data.info()
```

**4. Deleting the 'Status' and 'Unnamed1' columns as these have all values null**

```python
df_sales_data.drop(['Status' , 'unnamed1'] , axis = 1 , inplace = True)
df_sales_data
```

**5. Checking for any null values within the dataframe**

```python
df_sales_data.isnull().sum()
```

**6. Removing the null values**

```python
df_sales_data.dropna(inplace = True)
df_sales_data
```

**7. Cross - checking for null values**

```python
df_sales_data.isnull().sum()
```

*Exploratory Data Analysis*

**1. No. of Orders v/s Gender**

```python
df_sales_data_gender = df_sales_data['Gender'].value_counts().plot(kind = 'bar' , color = '#7e0315')

for bars in df_sales_data_gender.containers:
    df_sales_data_gender.bar_label(bars)
```

**2. Total Sales v/s Gender**

```python
df_sales_data_gender1 = df_sales_data.groupby(['Gender'] , as_index = False)['Amount'].sum().sort_values(
    by = 'Amount' , ascending = False).plot(kind = 'bar' , color = '#ff5a36' , x = 'Gender' , y = 'Amount')
```

*KT: Most of the buyers are Females and the purchasing power of Females is also greater than Males*

**3. Total no. of Orders v/s Age Group**

```python
df_age = sns.countplot(data = df_sales_data , x = 'Age Group' , hue = 'Gender')
for bars in df_age.containers:
    df_age.bar_label(bars)
```

**4. Total Sales v/s Age Group**

```python
df_age1 = df_sales_data.groupby(['Age Group'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False)
sns.barplot(x = 'Age Group' , y = 'Amount' , data = df_age1)
```

*KT: Most of the buyers are Females whose age is between 26 and 35*

**5. Top 10 States w.r.t No. of Orders**

```python
sales_state = df_sales_data.groupby(['State'] , as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state , x = 'State' , y = 'Orders')
```

**6. Top 10 States w.r.t Total Sales**

```python
sales_state = df_sales_data.groupby(['State'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state , x = 'State' , y = 'Amount')
```

*KT: The highest number of Orders and Total Sales are from Uttar Pradesh, Mahaeashtra, and Karnataka*

**7. Total Orders v/s Occupation**

```python
sales_state = df_sales_data.groupby(['Occupation'] , as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending = False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state , x = 'Occupation' , y = 'Orders')
```

**8. Total Sales v/s Occupation**

```python
sales_state = df_sales_data.groupby(['Occupation'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state , x = 'Occupation' , y = 'Amount')
```

*KT: The highest number of Orders and Total Sales is from IT Sector, Healthcare, and Avaiation sectors*

**9. Total Orders v/s Product Category**

```python
sales_product = df_sales_data.groupby(['Product_Category'] , as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_product , x = 'Product_Category' , y = 'Orders')
```

**10. Total Sales v/s Product Category**

```python
sales_product = df_sales_data.groupby(['Product_Category'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_product , x = 'Product_Category' , y = 'Amount')
```

*KT: The highest number of Orders are for Clothing & Apparel and highest Sales is for Food*

**CONCLUSION:**

*Females aged between 26 - 35 from Uttar Pradesh, Maharashtra, and Karnataka working in IT, healthcare, and Aviation are more likely to buy products like Food, Clothing, and Electronics*
